## Data visualization

The main library for visualizing data is matplotlib. You can use the `plot` function to visualize points. Each point can be represented by different markers:

- `o` is a circle
- `s` is a square
- `^` is a triangle
- `p` is a pentagon
- `*` is a star
- `+` is a plus
- `x` is a cross
- `D` is a diamond
- `h` is a hexagon
- `8` is an octagon


```python
import matplotlib.pyplot as plt

# visualize a single point with a marker (o is a circle, s is a square, ^ is a triangle,p is a pentagon, * is a star, + is a plus, x is a x, D is a diamond, h is a hexagon, 8 is a octagon)
plt.plot(3, 4, "o", color="red")
plt.plot(7, 8, "s", color="green", markersize=10)
plt.plot(5, 5, "^", color="blue", markersize=5)
```

plt.show ends the current plot

```python
plt.plot(3, 4, "o", color="red")
plt.show()
plt.plot(7, 8, "s", color="green", markersize=10)
```

```python
# plot with a legend
plt.plot(3, 4, "o", color="red", label="point 1")
plt.plot(7, 8, "s", color="green", markersize=10, label="point 2")
plt.legend()
```
