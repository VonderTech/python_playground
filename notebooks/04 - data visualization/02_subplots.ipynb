```python
import matplotlib.pyplot as plt
import numpy as np

plt.subplot(1, 2, 1)
plt.plot(np.random.randn(10), "bo-")

plt.subplot(2, 2, 2)
plt.plot(np.random.randn(10), "ro-")
plt.subplot(2, 2, 4)
plt.plot(np.random.randn(10), "go-")
```

```python
# with sublots

fig, ax = plt.subplots(2, 2)
ax[0, 0].plot(np.random.randn(10), "bo-")
ax[0, 0].plot(np.random.randn(10), "go-")
ax[0, 1].plot(np.random.randn(10), "ro-")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a figure with a 3x3 grid of subplots
fig, ax = plt.subplots(3, 3)


def plot_subplots_with_flatten():
    """
    Plot random data on each subplot using the flatten method.

    The flatten method is used to iterate over each subplot in a flattened
    1-dimensional array. This allows us to easily apply operations or plot
    data on each individual subplot.

    In this example, we generate random data using numpy's random.randn function
    and plot it on each subplot using the 'bo-' format (blue circles connected
    by lines).

    Note: The flatten method returns a 1-dimensional array of subplots, which
    is useful when we want to perform operations on each individual subplot.

    Returns:
        None
    """
    for i in ax.flatten():
        i.plot(np.random.randn(10), "bo-")


# Call the function to plot the subplots with flatten
plot_subplots_with_flatten()
```
