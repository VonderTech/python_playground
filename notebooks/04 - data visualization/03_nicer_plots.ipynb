```python
import matplotlib.pyplot as plt
import numpy as np

numbers = np.linspace(-10, 10, 50)

plt.plot(numbers, numbers, label="y = x")
plt.plot(numbers, numbers**2, label="y = x^2")
plt.plot(numbers, numbers**3, label="y = x^3")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Graphs of y = x, y = x^2, and y = x^3")
plt.xlim(-10, 10)
plt.gca().set_aspect(0.01)
plt.show()
```
