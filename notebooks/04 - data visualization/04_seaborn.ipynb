```python
from turtle import title
import numpy as np
import matplotlib.pyplot as plt
import pandas
import seaborn as sns

num = 200
dis = np.zeros((num, 2))

dis[:, 0] = np.linspace(0, 10, num) + np.random.randn(num)
dis[:, 1] = dis[:, 0] ** 2 + np.random.randn(num) * 10

dataframe = data = pandas.DataFrame(dis, columns=["first", "second"])


sns.jointplot(data=dataframe, x="first", y="second", color="b")
plt.title("Joint plot")
plt.show()

sns.regplot(data=dataframe, x="first", y="second")
plt.title("Regression plot")
plt.show()
```
