```python
from re import M
import matplotlib.pyplot as plt
import numpy as np

m = 3
n = 10

M = np.random.randint(10, size=(m, n))
print(M)

for i in range(m):
    for j in range(n):
        if M[i, j] > 5:
            # pay attention here to the swapping of i and j. This is because columns are x and rows are y
            plt.text(j, i, M[i, j], ha="center", va="center", color="black")
        else:
            plt.text(j, i, M[i, j], ha="center", va="center", color="white")

plt.imshow(M, interpolation="nearest")
plt.colorbar()
plt.show()
```

```python
import imageio.v3 as iio

url = "https://upload.wikimedia.org/wikipedia/commons/5/56/E%3DMC%5E2_%287852234992%29.jpg"
img = iio.imread(url)

plt.imshow(img)
plt.axis = "off"
plt.show()

plt.hist(
    img.ravel(),
    bins=1000,
    range=(0, 255),
    density=True,
    histtype="step",
    color="black",
)
plt.show()
```
